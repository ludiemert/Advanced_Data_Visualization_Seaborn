```python
# fazer uma analise desses dados com essa lib
# passar uma coluna que quer fazer um plot de distribuicao, passando uma coluna de DF

# analise de qto as pessoas gastam
sns.displot(tips['total_bill'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.displot(tips['total_bill'], kde=True)
# A curva que você quer é uma curva de densidade (KDE), não apenas histograma
# Por padrão, sns.displot() mostra histograma.
# Se quiser a curva suave de estatística, ative o kde=True:

plt.show()
```

```python
# definir qtas subdivisoes eu quero no grafico
sns.displot(tips['total_bill'], kde=True, bins=300) # atualimente tem umas 10
```

```python
# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas

# analise com o DF no fim
sns.jointplot(x='total_bill', y='tip', data=tips)

# grafico de distribuicao conjunta
# grafico scater
```

```python
# outro tipo de grafico, grafico de regrecao
# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas
# grafico de distribuicao conjunta
# analise com o DF no fim

sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')

```

```python
# outro tipo de grafico, grafico de hexadecimal de analise

# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas
# grafico de distribuicao conjunta
# analise com o DF no fim

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# grafico que procura todas as variaveis numericas, variaveis numericas

sns.pairplot(tips) # visao geral dos dados
```

```python
# separar os dados ex text e num, sex fem e sex masc

sns.pairplot(tips, hue='sex')
```

```python
# usar paleta de cores, pegar cor quente e cor fria

sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# grafico forma de ver concentracao de dados
sns.rugplot(tips["total_bill"])
```

```python
# mix de graficos
sns.kdeplot(tips["total_bill"])
```

```python
# add dois dados e dois graficos juntos
sns.kdeplot(tips["total_bill"])
sns.rugplot(tips["total_bill"])
```
