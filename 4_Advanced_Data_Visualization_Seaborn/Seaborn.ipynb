```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.displot(tips['total_bill'], kde=True)
# A curva que você quer é uma curva de densidade (KDE), não apenas histograma
# Por padrão, sns.displot() mostra histograma.
# Se quiser a curva suave de estatística, ative o kde=True:

plt.show()
```

```python
# definir qtas subdivisoes eu quero no grafico
sns.displot(tips['total_bill'], kde=True, bins=300) # atualimente tem umas 10
```

```python
# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas

# analise com o DF no fim
sns.jointplot(x='total_bill', y='tip', data=tips)

# grafico de distribuicao conjunta
# grafico scater
```

```python
# outro tipo de grafico, grafico de regrecao
# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas
# grafico de distribuicao conjunta
# analise com o DF no fim

sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')

```

```python
# outro tipo de grafico, grafico de hexadecimal de analise

# analisar duas variaveis ao mesmo tempo
# permite que crie duas variacoes e uma terceira com uma relacao entre elas
# grafico de distribuicao conjunta
# analise com o DF no fim

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# grafico boxplot

import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# carregar dataset
# sns.boxplot(x="day", y="total_bill", data=tips, palette='rainbow')

sns.boxplot(x="day", y="total_bill", data=tips, hue="day", palette='rainbow', legend=False)

plt.show()
# grafico de distribuicao com percentil de dados, ele distrib onde os dados se distribuem
# a linha do meio eh a mediana nao a media
# mostra as pessoas que temos interesse em analisar
# ele mostra como as contas estao distrib por dia da semana
```

```python
# outra forma de mostrar os dados via horizontal do boxplot
sns.boxplot(y="day", x="total_bill", data=tips, hue="day", palette='rainbow', )
```

```python
# outra forma de mostrar os dados via horizontal do boxplot
sns.boxplot(y="day", x="total_bill", data=tips, hue="sex", palette='rainbow', legend=True, orient="h")
```

```python
# outra forma de mostrar os dados via horizontal do boxplot
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette='rainbow')
```

```python
# other grafic violinplot
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", palette='rainbow') # o  hue="sex" faz ligacao entre homens e mulheres


# ve estimativa da carga de distribuicao
```

```python
# usando o method split condensa informacao e economiza espaco
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", palette='rainbow', split=True) # o  hue="sex" faz ligacao entre homens e mulheres
```

```python
# swarmplot => agrupa valores de maneira real
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
# comercar a correlacionar variaveis

sns.lmplot(x="total_bill", y="tip", data=tips)


# cria um grafico de regressao entre os dois valores
```

```python
# comercar a correlacionar variaveis, fazer comparacao entre valores

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex")


# cria um grafico de regressao entre os dois valores e comparacao
```

```python
# outros atributos para customizar essas informacoes, com marcacao diferente (bolinha e triangulo)

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['o', 'v'])
```

```python
# outros atributos para customizar essas informacoes, com marcacao diferente (x e triangulo)

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['x', 'v'])
```

```python
# mudar o tamanho da informacao
# outros atributos para customizar essas informacoes, com marcacao diferente (bolinha e triangulo)

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['o', 'v'], scatter_kws={'s': 100})
```

```python
# multiplas regressoes do time em varias regressoes

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="time")
# cria 2 graficos de almoco e jantar
```
